Задание 1 Переведите содержимое файла purchase_log.txt в словарь purchases вида: {'1840e0b9d4': 'Продукты', ...}

```python
purchases={}

with open('purchase_log.txt', 'r', encoding='utf-8') as file: #берем кодировку utf-8 что бы ру язык не глючил.
    next(file) # Пропустим первую строку
    for line in file: # Разбиваем строку на список слов
        words = line.strip('{}').split()
        user_id = words[1].strip('{,"}')
        category = words[3].strip('{,"}') # так и не понял почему пришлось чистить каждую строку, но в итоге удалось.
        #выделили по заданию второе и четвертое слово из каждой строки. Осталось добавить их в словарь
        purchases[user_id] = category


print(purchases)
```

Задание 2 Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки. Если покупка была, сам файл visit_log.csv изменять не надо. Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории. Учтите условия для данных: ● содержимое purchase_log.txt помещается в оперативную память компьютера; ● содержимое visit_log.csv — нет, используйте только построчную обработку этого файла

```python
purchases={}
#подготовим словарь как в задании 1
with open('purchase_log.txt', 'r', encoding='utf-8') as file: #берем кодировку utf-8 что бы ру язык не глючил.
    next(file) # Пропустим первую строку
    for line in file: # Разбиваем строку на список слов
        words = line.strip('{}').split()
        user_id = words[1].strip('{,"}')
        category = words[3].strip('{,"}') # так и не понял почему пришлось чистить каждую строку, но в итоге удалось.
        #выделили по заданию второе и четвертое слово из каждой строки. Осталось добавить их в словарь
        purchases[user_id] = category

with open('visit_log.csv', 'r', encoding='utf-8') as visit_file:
    with open('funnel.csv', 'w', encoding='utf-8') as funnel_file:

        header = visit_file.readline().strip()
        funnel_file.write(f"{header},category\n") #считаем заголовок из визитов и запишем в фаннел.

        # Поочереди обработаем каждую линию фаила visit_log и выпишем покупки с указаниями категорий убрав лишние символы.
        for line in visit_file:
            parts = line.strip().split(',')
            if len(parts) > 0: # Уберем пустые строки
                user_id = parts[0] # Примем то что user id расположен в первой колонке и будем это использовать

                # ищем категорию user_id через гет, если она есть то пишем ее.
                category = purchases.get(user_id)
                if category:
                    funnel_file.write(f"{line.strip()},{category}\n")

import pandas as pd
data = pd.read_csv('funnel.csv')
print(data)
```
